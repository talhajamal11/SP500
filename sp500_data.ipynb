```python
Unique_Stocks = Stocks['Symbol'].unique()
Unique_Stocks[7]
```

```python
Col = ['Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']
Daily_Pricing = pd.DataFrame(columns=Col)
```

```python
for i in range(0, len(Unique_Stocks)):
    Tic = Unique_Stocks[i]
    Tic_data = yf.download(Tic)
    Tic_data['Ticker'] = Tic
    Tic_data = Tic_data[['Ticker', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
    Daily_Pricing = Daily_Pricing.append(Tic_data)
```
